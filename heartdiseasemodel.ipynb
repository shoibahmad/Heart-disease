```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(Y)
```

```python
print(X)
```

```python
#splitting the data  into training data & test Data
```

```python
X_train, X_test, Y_train,Y_test= train_test_split(X, Y,test_size=0.2, stratify = Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#modeltraining logistic regression 
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
#trainingt the LogisticRegression model with training data
```

```python
model.fit(X_train, Y_train)
```

```python
#model evaluation
```

```python
#Accuracy score
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

```python
# Calculate accuracy
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)

# Prepare data for plotting
labels = ['Training Data']
accuracies = [training_data_accuracy * 100]  # Convert to percentage

# Create the plot
plt.figure(figsize=(6, 4))
plt.bar(labels, accuracies, color='skyblue')
plt.ylim(0, 100)  # Set y-axis limit to 0-100%
plt.ylabel('Accuracy (%)')
plt.title('Model Training Accuracy')

# Annotate the accuracy percentage on the bar
for i, acc in enumerate(accuracies):
    plt.text(i, acc + 1, f'{acc:.2f}%', ha='center', fontsize=12)

# Show the plot
plt.show()
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ',test_data_accuracy)
```

```python
import pandas as pd

# Input data
input_data = (54, 0, 1, 132, 288, 1, 0, 159, 1, 0, 2, 1, 2)

# Feature names
feature_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 
                 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']

# Convert input data to a DataFrame
input_data_df = pd.DataFrame([input_data], columns=feature_names)

# Choose prediction mode: 'precision', 'recall', or 'default'
mode = 'precision'  # Change to 'recall' or 'default' as needed

# Set threshold based on desired mode
if mode == 'precision':
    threshold = 0.7  # High threshold reduces false positives
elif mode == 'recall':
    threshold = 0.3  # Low threshold reduces false negatives
else:
    threshold = 0.5  # Default threshold

# Predict probability of class 1 (heart disease)
proba = model.predict_proba(input_data_df)[0][1]

# Output result
print(f"Predicted probability of heart disease: {proba:.2f}")
if proba >= threshold:
    print("Prediction: The person has heart disease.")
else:
    print("Prediction: The person does not have heart disease.")
```
